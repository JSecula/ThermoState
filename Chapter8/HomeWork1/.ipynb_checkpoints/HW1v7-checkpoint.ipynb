## Problem Statement

Water is the working fluid in an ideal Rankine cycle. At the inlet to the turbine, the quality of the vapor is $x_1 =$ 1.0  and the pressure is $p_1 =$ 16.0 MPa. The condenser pressure is $p_2 =$ 8.0 kPa. The mass flow rate of steam entering the turbine is $\dot{m} =$ 120.0 kg/s. Determine:

1. the net power developed, in kW
2. the rate of heat transfer to the steam passing through the boiler, in kW
3. the thermal efficiency of the cycle
4. the mass flow rate of cooling water, in kg/s, if the cooling water undergoes a temperature increase from $T_{cw,1} =$ 18.0 celsius to $T_{cw,2} =$ 36.0 celsius with negligible pressure change in passing through the condenser and with constant specific heat evaluated at $p =$ 1 atm and $T_{avg} =$ 27.0 celsius


```python
# Engineering Model
---
## 1. Steady State
## 2. 1-D flow
## 3. Ignor any possible stray heat tranfer
## 4. KE and PE are negligable

from thermostate import State, Q_, units
print ('Jason Secula')
print ('Homework One')
print ('     ')
#Homework 1-1 Question
#Find Net Power Developed in kW
#Rate of Heat Transfer through boiler
#Thermal Efficency of Cycle
#MassFlowRate of cooling water  in kg/s after temp change.

substance = 'water'

#State One Adiabatic Expansion Through a TURBINE (1-2)
x_1 = Q_(1.0, 'dimensionless') #Given Vapor
p_1 = Q_(16.0, 'MPa') #Given Pressure into the Turbine
mdot = Q_(120.0, 'kg/s') #Given MassFlowRate entering Turbine

st_1 = State(substance, p=p_1, x=x_1)
print('Program: Start at Turbine ')

h_1 = st_1.h #Get and Store Enthalpy (h) as a Variable from State 1
s_1 = st_1.s
T_1 = st_1.T#Adiabatic (No Change in heat)
s_2 = s_1

#State Two Isobaric Compression through a Condenser (2-3)
p_2 = Q_(8.0, 'kPa') #Given Pressure of the Condenser
st_2 = State(substance, p=p_2, s=s_2) #Isobaric (No change in pressure)
print('Program: End of Turbine, Start of Condenser ')

h_2 = st_2.h #Get and Store Enthalpy (h) as a Variable from State 2
p_3 = p_2 #Isobaric (No change in pressure)
T_2 = st_2.T
T_3 = T_2 #Adiabatic process, temperature is constant through Pump.
x_3 = Q_(0.0, 'dimensionless') #the quality is at zero because state 3 is on the saturated line. 

#State Three Adiabatic Compression through the PUMP (3-4)
st_3 = State(substance, p=p_3, x=x_3)
print('Program: End of Condensor, Start of Pump')

s_4 = st_3.s
h_3 = st_3.h
p_4 = st_3.p #Set the Pressure4 equal to pressure from state 3.
print (h_3)

#State Four Isobaric Expansion through the Boiler (4-1)
st_4 = State(substance,s=s_4,p=p_4)
print('Program: End of Pump, Start of Boiler')
print ('     ')
h_4 = st_4.h
print (h_4)

W_t = (mdot*(h_1-h_2))
print (' ** The Work done by the Turbine was ')
print(W_t.to('kW'))
#^Unable to convert within print statement, or at least my syntax was wrong.
#^^I kept getting a dimentionality error, or it was trying to convert to dimensionless.


W_p = (mdot*(h_4-h_3))
print (' ** The Work done by the Pump was ')
print (W_p.to('kW'))
#^I feel like there should be a easier way to convert values, but I dont know how.
print ('     ')

W_Net = (W_t + W_p)
print ('1. The Net Work Output ')
print (W_Net.to('kW'))
#I wanted to print the String and the variable in the same line, but I kept getting an error.
print ('     ')

#Equation 8.4 on page 447 of the text, shows Q_in/mdot=h_1-h_4
Qin = mdot*(h_1-h_4)
print ('2. Heat transfered into the boiler ')
print (Qin.to('kW'))
print ('     ')

#Page 451 of text has a similar problem. 
n = ((W_Net)/Qin)*100   #(WdotT+(-WdotP))/Qin, or [(h_1-h_2)+(h_4-h_3)]/(h_1-h_4)
print ('3. The Thermal Efficiency of this cycle is ')
print (n)
print ('     ')

#NumberFour, mdotCW = (mdot*(h_2-h_3))/(h_3b-h_2b) page 452
Tcw_1 = Q_(18.0, 'degC')#Given Cooling Water 1
p_5 = Q_(1.0, 'atm') #Given Constant pressure for Cooling water
st_5 = State(substance, T=Tcw_1, p=p_5)
h_2b = st_5.h #h_5 = 75637.71261772985 j/kg

#Cooling Temp increases from 18C to 36C with p = 1 atm 
Tcw_2 = Q_(36.0, 'degC') #Given Cooling Water 2
st_6 = State(substance, T=Tcw_2, p=p_5)
h_3b = st_6.h #h_6 = 150899.11328385773j/kg

T_avg = Q_(27.0, 'degC') #Given Average of Cooling Water 2>>3
st_7 = State(substance, T=T_avg, p=p_5)
h_avg = st_7.h #average enthalpy = 113281.99139244082j/kg

mCWavg = (mdot*(h_2-h_3))/(h_avg)
print(' ** 4. The average Mass flow rate of the Cooling water through the Condenser is ')
print(mCWavg.to('kg/s')) #Using the average is much lower...
print ('     ')

mCW = (mdot*(h_2-h_3))/(h_3b-h_2b)
print('4. The Mass flow rate of the Cooling water through the Condenser is ')
print(mCW.to('kg/s'))
print ('     ')
print ('     ')
print ('Program Complete.')
```
