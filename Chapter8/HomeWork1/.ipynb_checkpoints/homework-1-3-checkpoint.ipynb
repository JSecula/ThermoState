## Definitions

```python
substance = 'water'
Wdot_net = Q_(210.0, 'MW')           #The total work out
T_1 = Q_(500.0, 'degC')              #The Temperature going into the Turbine. 
p_1 = Q_(10.0, 'MPa')                #The Pressure at state one.
p_2 = Q_(10.0, 'kPa')                #The Pressure at state two.
eta_t = eta_p = Q_(85.0, 'percent')  #Thermal efficentcy is 85%
```

## Problem Statement

Consider a 210.0 MW steam power plant that operates on an ideal superheated Rankine cycle. Steam enters the turbine at 10.0 MPa and 500.0 celsius and is cooled in the condenser at a pressure of 10.0 kPa. Determine:

1. the quality of the steam at the exit of the turbine
2. the thermal efficiency of the cycle
3. the mass flow rate of the steam
4. 1.-3. again, if the isentropic efficiencies of the pump and turbine are both 85.0  percent.
