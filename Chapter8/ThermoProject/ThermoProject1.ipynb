```python
from thermostate import State, Q_, units
import matplotlib.pyplot as plt
from math import pi

units.default_format = '~H'
substance = 'water'
#substance = 'ammonia'
eta_t = eta_p = Q_(80.0, 'percent') 
# I would like to have mostly efficient system...

print('Jason Secula')
print('Piston Power')
print('02/19/2017')

#FactorEngine Specifications.
power = Q_(6.5, 'hp') #HorsePower From LawnMowerEngine
CArea = Q_((pi*(2.69/2)),'in**2') #Cylinder Cross Sectional Area
CVolm = Q_(11.57,'cc') #CC Listed on  LawnMowerEngine
EVolm = Q_(11.594,'in**3') #Calculated Volume from Bore and Stroke
Torq = Q_(6.7,'lbf') #Torque on the Engine in Foot Pound
Erpm = Q_(2600,'rpm') #According to Briggs and Strattons Website, engines are gages to 2600 rpm
FWRad = (0.62/2) #Radius of the Cam Shaft at the Fly Wheel (The Thinest Part)

#Maximum Force we can turn the flywheel without the camshaft breaking.
J = ((pi*(FWRad**4))/2) #Polar Moment of Inertia
Twist = (power/Erpm) #Twisting Moment
ToaMax = ((Twist*FWRad)/J)
#Veloicity = Integral (volume)dt from 0 rpm to 1 rpm. 
#(volume*t) * (t^3)/3 where t varies from 0 to 1 Hz or 60 seconds. 
Velocity = Q_(((EVolm*60)*((60**3)/3)),'in/sec')
print(' The Maximum shear on the crank shaft is = {}'.format(ToaMax))

#ValveTimeing Courtesy of HiperMath.com
SpdSnd = Q_(67547.4,"ft/minute").to('in/minute') #Speed of Sound
AvgOpenIntake = ((0.327*EVolm*Erpm*12)/(SpdSnd * 12)) #The average Open Valve intake Area
print(' The Intake Valve Area is = {}'.format(AvgOpenIntake)) #ThermoState Didnt simply
#The Units are weird, I'm not sure if this is right. 

#Inlet Nozzle
OutletB = Q_((pi*((1/32)/2)),'in**2')
NzlFarce = Q_(100,'lbf') #We need at least 65.7 lbin^2 to turn the turbine
p_1 = (NzlFarce/OutletB).to('psi')

T_1 = Q_(950.0, 'degF')
rhoW = Q_(1.4702, 'lb/ft**3').to('lb/in**3') #Found Density after setting temperature and pressure.
Mdot = ((rhoW*Velocity*CArea)).to('kg/sec') #Mdot = 0.041681 in^2/lb

#State One Super Heater Vapor into turbine. 
st_1 = State(substance, T=T_1, p=p_1)
s_1 = st_1.s
print('When StateOne = {}'.format(st_1.phase) + ', The Entropy = {}'.format(st_1.s))
print('and When StateOne = {}'.format(st_1.phase) +', The Enthalpy  is = {}'.format(st_1.h))
print('and when StateOne = {}'.format(st_1.phase) +', The Temperature = {}'.format(st_1.T))
print('and when StateOne = {}'.format(st_1.phase) +', The Pressure = {}'.format(st_1.p))
print('and when StateOne = {}'.format(st_1.phase) +', The Volume is = {}'.format(st_1.v))
print('and When StateOne = {}'.format(st_1.phase) +', The Internal Energy  is = {}'.format(st_1.u))
print('and when StateOne = {}'.format(st_1.phase) +', The Quality is = {}'.format(st_1.x))
print('and when StateOne = {}'.format(st_1.phase) +', The Volume is = {}'.format(st_1.v))
print('                        ')

#Outlet Diffuser
InletC = Q_((pi*((.25)/2)),'in**2')
Difzar = Q_(57,'lbf')
p_2 = (Difzar/InletC).to('psi')

#State TwoS, Exiting the Turbine, and entering the condenser
s_2 = st_1.s
s_2s = st_1.s
st_2s = State(substance, p=p_2, s=s_2s)
print('When StateOne = {} '.format(st_2s.phase) +  ', The Entropy = {}'.format(st_2s.s))
print('and When StateTwoS is  = {}'.format(st_2s.phase) +', The Enthalpy  is = {}'.format(st_2s.h))
print('and when StateTwoS is  = {}'.format(st_2s.phase) +', The Temperature = {}'.format(st_2s.T))
print('and when StateTwoS is  = {}'.format(st_2s.phase) +', The Pressure = {}'.format(st_2s.p))
print('and when StateTwoS is = {}'.format(st_2s.phase) +', The Volume is = {}'.format(st_2s.v))
print('and When StateTwoS is  = {}'.format(st_2s.phase) +', The Internal Energy  is = {}'.format(st_2s.u))
print('and when StateTwoS is  = {}'.format(st_2s.phase) +', The Quality is = {}'.format(st_2s.x))
print('                        ')
T_2s = st_2s.T

#StateThree, Leaving the Condenser, entering the 
p_3 = st_2s.p
x_3 = Q_(0.0, 'dimensionless')
st_3 = State(substance, p=p_3, x=x_3)
print('When StateThree =  {}'.format(st_3.phase) + ', The Entropy = {}'.format(st_3.s))
print('and when StateThree =  {}'.format(st_3.phase) +', The Enthalpy  is = {}'.format(st_3.h))
print('and when StateThree =  {}'.format(st_3.phase) +', The Temperature = {}'.format(st_3.T))
print('and when StateThree =  {}'.format(st_3.phase) +', The Pressure = {}'.format(st_3.p))
print('and when StateThree =  {}'.format(st_3.phase) +', The Volume is = {}'.format(st_3.v))
print('and when StateThree =  {}'.format(st_3.phase) +', The Internal Energy  is = {}'.format(st_3.u))
print('and when StateThree =  {}'.format(st_3.phase) +', The Quality is = {}'.format(st_3.x))
print('                        ')

#StateFourS
s_4s = st_3.s
p_4 = st_1.p
st_4s = State(substance, p=p_4, s=s_4s)
print('When StateFourS is  =   {}'.format(st_4s.phase) + ', The Entropy = {}'.format(st_4s.s))
print('and When StateFourS is  =  {}'.format(st_4s.phase) +', The Enthalpy  is = {}'.format(st_4s.h))
print('and when StateFourS is  =  {}'.format(st_4s.phase) +', The Temperature = {}'.format(st_4s.T))
print('and when StateFourS is  =  {}'.format(st_4s.phase) +', The Pressure = {}'.format(st_4s.p))
print('and when StateFourS is  =  {}'.format(st_4s.phase) +', The Volume is = {}'.format(st_4s.v))
print('and When StateFourS is  =  {}'.format(st_4s.phase) +', The Internal Energy  is = {}'.format(st_4s.u))
print('and when StateFourS is  =  {}'.format(st_4s.phase) +', The Quality is = {}'.format(st_4s.x))
print('                        ')

Wdot_t = ((st_1.h - st_2s.h)*Mdot) #Work Done By the Turbine
print('The Work Done by the Turbine = {}'.format(Wdot_t))
Wdot_p = ((st_3.h - st_4s.h)*Mdot)  #Work Done by the Pump
print('The Work Done by the Pump = {}'.format(Wdot_p))
Qdot_in = ((st_1.h - st_4s.h)*Mdot) #Heat put into the system
print('The Heat into the System = {}'.format(Qdot_in))
print('                        ')
print('                        ')
print(' Now Calculating the system as if it was 75% efficiency')
print('                        ')
print('                        ')
#With 80% Thermal Efficentcy

h_2 = eta_t*(st_2s.h - st_1.h) + st_1.h
x_2 = st_2s.x
#State Two, Exiting the Turbine, and entering the condenser
st_2 = State(substance, p=p_2, h=h_2)
print(' When StateTwo = {} '.format(st_2.phase) +  ', The Entropy = {}'.format(st_2.s))
print(' and When StateTwo is =  {}'.format(st_2.phase) +', The Enthalpy  is = {}'.format(st_2.h))
print(' and when StateTwo is =  {}'.format(st_2.phase) +', The Temperature = {}'.format(st_2.T))
print(' and when StateTwo is =  {}'.format(st_2.phase) +', The Pressure = {}'.format(st_2.p))
print(' and when StateTwo is  =  {}'.format(st_2.phase) +', The Volume is = {}'.format(st_2.v))
print(' and When StateTwo is =  {}'.format(st_2.phase) +', The Internal Energy  is = {}'.format(st_2.u))
print(' and when StateTwo is =  {}'.format(st_2.phase) +', The Quality is = {}'.format(st_2.x))
print('                        ')

#State Four
h_4 = (st_4s.h - st_3.h)/eta_p + st_3.h
st_4 = State(substance, p=p_4, h=h_4)
print(' When StateFour is = {}'.format(st_4.phase) + ',  The Entropy = {}'.format(st_4.s))
print(' and When StateFour is = {}'.format(st_4.phase) +', The Enthalpy  is = {}'.format(st_4.h))
print(' and when StateFour is = {}'.format(st_4.phase) +', The Temperature = {}'.format(st_4.T))
print(' and when StateFour is = {}'.format(st_4.phase) +', The Pressure = {}'.format(st_4.p))
print(' and when StateFour is  =  {}'.format(st_4s.phase) +', The Volume is = {}'.format(st_4s.v))
print(' and When StateFour is = {}'.format(st_4.phase) +', The Internal Energy  is = {}'.format(st_4.u))
print(' and when StateFour is = {}'.format(st_4.phase) +', The Quality is = {}'.format(st_4.x))
print('                        ')

Wdot_t_mdot = (st_1.h - st_2.h) #Calculating Mdot, Work by Turbine
print(' The Work Done by the Turbine/mdot = {}'.format(Wdot_t_mdot))
Wdot_p_mdot = st_3.h - st_4.h #Calculating Mdot, Work by pump
print(' The Work Done by the Pump/mdot = {}'.format(Wdot_p_mdot))
Qdot_in_mdot = (st_1.h - st_4.h) #Calculating Mdot, Heat transfer by Q_in
print(' The Heat into the System over mdot = {}'.format(Qdot_in_mdot))
print('                        ')

Wnet1 = (Wdot_t+Wdot_p)
print(' The Net Power Created is = {}'.format(Wnet1))
Wnet2 = (Wdot_t_mdot+Wdot_p_mdot)
print(' The Net Power Created with an efficiency of 80% = {}'.format(Wnet2))
print('                        ')

eta_T1 = ((Wdot_t + Wdot_p)/Qdot_in)
print(' The Thermal Efficiency is = {}'.format(eta_T1))
eta_T2 = ((Wdot_t_mdot + Wdot_p_mdot)/Qdot_in_mdot)
print(' The Thermal Efficiency Calculated with a 80% efficency = {}'.format(eta_T2))
print('                        ')

#plt.plot([st_1.T, st_2s.T, st_3.T, st_4s.T, st_2.T, st_4.T],[st_1.s, st_2s.s, st_3.s, st_4s.s, st_2.s, st_4.s])
#plt.title('Temp vs. Entropy')
#plt.ylabel('Temp')
#plt.xlabel('Entropy')
#plt.show() 
```
