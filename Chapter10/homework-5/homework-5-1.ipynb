```python
from thermostate import State, Q_, units
from math import pi
substance = 'ammonia'           #Given Info
Qdot_in = Q_(127.0, 'kW')
#_____________________________________________________________________ Line
print('Jason Secula'.center(85))
print(' Applied Thermodynamics [ME 2234] '.center(85))
print(' Professor Bryan W. Weber '.center(85))
print(' Homework 5.1 '.center(85))
print('           ')
#_____________________________________________________________________ Line
#__________________________State 1____________________________________
#_____________________________________________________________________
x_1 = Q_(1.0, 'dimensionless')
T_1 = Q_(-22.0, 'degC')
st_1 = State(substance, T=T_1, x=x_1)
#_____________________________________________________________________ Line
#__________________________State 2____________________________________
#_____________________________________________________________________
T_2 = Q_(180.0, 'degC')
p_2 = Q_(16.3, 'bar')
st_2 = State(substance, T=T_2, p=p_2)
#_____________________________________________________________________ Line
#__________________________State 3____________________________________
#_____________________________________________________________________
x_3 = Q_(0.0, 'dimensionless')
st_3 = State(substance, x=x_3, p=p_2)
#_____________________________________________________________________ Line
#__________________________State 4____________________________________
#_____________________________________________________________________
st_4 = State(substance, s=st_3.s, T=st_1.T)
#_____________________________________________________________________ Line
#__________________________State s____________________________________
#_____________________________________________________________________
#h_2s = ((Wc/mdot)+st_1.h).to('J/kg')
h_4s = st_3.h
st_2s = State(substance, s=st_1.s, p=p_2 )
st_4s = State(substance, p=st_1.p, h=st_3.h)
#_____________________________________________________________________ Line
#__________________________Calculations ______________________________
#_____________________________________________________________________
mdot = (Qdot_in/(st_1.h-st_4.h)).to('kg/s')
print ('1. Mass flow rate                   =    {}'.format(mdot))
Wc = (mdot*(st_1.h-st_2.h)).to('kW')
print ('2. Compressor Work input            =    {}'.format(Wc))
Wt = (mdot*(st_3.h-st_4.h)).to('kW')
print ('     The Trottle work               =    {}'.format(Wt))
print ('     The Evaporator Qin             =    {}'.format(Qdot_in))
Wcycle = Wc-Wt
print ('     The Total Work done            =    {}'.format(Wcycle))
print ('                    ')
β = (Qdot_in/(Wcycle))                                   #Eq found in Textbook.
β1 = (Qdot_in/(Wc))                                      #Eq given in class.
βmax = (st_1.T/(st_2.T-st_1.T))                          #Beta Max
print ('3. Preformance(β = Qin/Wc)          =    {}'.format(β1))
print ('     Preformance(β = Qin/Wcycle)    =    {}'.format(β))
print ('     (βmax = Tc/(TH-Tc))            =    {}'.format(βmax))
#__________^^^^My notes say this should be between 3 and 20^^^^_______
η_Comp = ((st_2s.h-st_1.h)/(st_2.h-st_1.h)).to('percent')
print ('4.  The Isentropic Efficiency(η_c) is ({}).'.format(η_Comp))
S = ((Wt/st_1.T)-(Wc/st_2.T)).to('kW/K')
print ('5.  The Rate of Entropy Production is ({}).'.format(S))
#https://en.wikipedia.org/wiki/Entropy_production
#^ 0 = Q_h/T_h - Q_c/T_c + S
#^ Power = (((T_h-T_c)/T_h)*Q_H)-T_c*S
# (Power - (((T_h-T_c)/T_h)*Q_H))) / -T_c = S
print ('                    ')
print ('                    ')
print(' Answers Using 2s'.center(85))
WSc = (mdot*(st_1.h-st_2s.h)).to('kW')
print ('2b. Compressor Work                     =    {}'.format(WSc))
print ('        The Trottle work                =    {}'.format(Wt))
print ('        The Evaporator Qin              =    {}'.format(Qdot_in))
Qout = (mdot*(st_2.h-st_3.h))
print ('        The Condensor Qout              =    {}'.format(Qout))
WScycle =((mdot*(st_1.h-st_2s.h))-(mdot*(st_3.h-st_4.h))).to('kW')
print ('        The Total Work(s) done          =    {}'.format(WScycle))
βs = ((st_1.h-st_4.h)/(st_1.h-st_2s.h))             #AbsoluteValue is above 3
βsMax = (st_1.T/(st_2.T-st_1.T))                    #Beta Max
print ('3b. Preformance(β = Qin/Wc)             =    {}'.format(βs))
print ('     (βmax = Tc/(TH-Tc))                =    {}'.format(βsMax))
S2 = ((Wc-(((st_2.T-st_1.T)/st_2.T)*Qout))/st_1.T).to('kW/K')
print ('5b.  The Rate of Entropy Production     =    {}'.format(S2))
```
