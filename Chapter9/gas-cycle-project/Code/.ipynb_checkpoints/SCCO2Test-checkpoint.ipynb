```python
from thermostate import State, Q_, units
from math import pi
#___________________________________
import matplotlib.pyplot as plt
import scipy.special as special
import numpy as np
#_____________________________________________________________________ Line
substance = 'AIR' #R-744 for CO2 did not work...
#_____________________________________________________________________ Line
#_____________________State 1, Into Compressor 1._____________________
#_____________________________________________________________________
p_1 = Q_(7.69231, 'MPa')
T_1 = Q_(305.15, 'K')
st_1 = State(substance, T=T_1, p=p_1)
#_____________________________________________________________________ Line
#_____________________State 2, Into Regenerator 2.____________________
#_____________________________________________________________________
p_2 = Q_(20, 'MPa')
T_2 = Q_(334.25, 'K')
st_2 = State(substance, T=T_2, p=p_2)
#_____________________________________________________________________ Line
#__________State 2s, Into Regenerator 1. with irrevsibilities.________
#_____________________________________________________________________
#Compressor is Isentropic: s_1 = s_2s
st_2s = State(substance, p=p_2, s=st_1.s)
#_____________________________________________________________________ Line
#_____________________State 3, Into Regenerator 1.____________________
#_____________________________________________________________________
p_3 = Q_(19.98868, 'MPa')
T_3 = Q_(431.14, 'K')
st_3 = State(substance, T=T_3, p=p_3)
#_____________________________________________________________________ Line
#_____________________State 3, Into Regenerator 1.____________________
#_____________________________________________________________________
st_3s = State(substance, s=st_1.s, p=p_3) 
#_____________________________________________________________________ Line
#_____________________State 4, Into Heat Exchanger._______[ Q_in ]____
#_____________________________________________________________________
p_4 = Q_(19.95795, 'MPa')
T_4 = Q_(669.69, 'K')
st_4 = State(substance, T=T_4, p=p_4)
#_____________________________________________________________________ Line
#_____________________State 4s, Into Heat Exchanger._______[ Q_in ]____
#_____________________________________________________________________
st_4s = State(substance, T=T_4, p=p_3)
#_____________________________________________________________________ Line
#_____________________State 5, Into Turbine_______________[ W_c/m. ]__
#_____________________________________________________________________
T_5 = Q_(823.15, 'K')
st_5 = State(substance, T=T_5, p=p_4)
#_____________________________________________________________________ Line
#_____________________State 5s, Into Turbine_______________[ W_c/m. ]__
#_____________________________________________________________________
p_5s = Q_(19.82795, 'MPa')
st_5s = State(substance, T=T_5, p=p_5s)
#_____________________________________________________________________ Line
#_____________________State 6, Back into Regenerator 1._______________
#_____________________________________________________________________
p_6 = Q_(7.90116, 'MPa')
T_6 = Q_(713.44, 'K')
st_6 = State(substance, T=T_6, p=p_6)
#_____________________________________________________________________ Line
#__________State 6s, Back into Regenerator 1. with irrevsibilities____
#_____________________________________________________________________
#Use Turbine is Adiabatic: s_5 = s_6s
st_6s = State(substance, p=p_6, s=st_5.s)
#_____________________________________________________________________ Line
#_____________________State 7, Back into Regenerator 2._______________
#_____________________________________________________________________
p_7 = Q_(7.81421, 'MPa')
T_7 = Q_(441.49, 'K')
st_7 = State(substance, T=T_7, p=p_7)
#_____________________________________________________________________ Line
#_____________________State 8, Split _________[ to HE and Comp2 ]_____
#_____________________________________________________________________
T_8 = Q_(342.74, 'K')
st_8 = State(substance, T=T_8, p=st_1.p)
#_____________________________________________________________________ Line
#_____________________State 8s, Split _________[ to HE and Comp2 ]_____
#_____________________________________________________________________
p_8s = Q_(7.70458, 'MPa')
st_8s = State(substance, T=T_8, p=p_8s)
#_____________________________________________________________________ Line
#_____________________Work Calculations ______________________________
#_____________________________________________________________________
Wd_cyc = Q_(85.5,'kJ/kg') #Given work output by the cycle. 
#________________________
Qd_in_md = (st_5.h-st_4.h)
Wd_t_md = (st_5.h-st_6.h)
Wd_c_md = ((st_2.h-st_1.h)+(st_8.h-st_3.h)) #Page 551 of Text
print('54 (Qdot_in/mdot)      = {}'.format(Qd_in_md))
print('56 (Wdot_t/mdot)       = {}'.format(Wd_t_md))
print('21+83 (Wdot_c/mdot)    = {}'.format(Wd_c_md))
print('           ')
Wd_t_md2 = (st_1.cp*(T_5-T_6))
print('56s (Wdot_t/mdot)   = {}'.format(Wd_t_md2))
Wd_c_md2 = ((st_2s.h-st_1.h)+(st_8s.h-st_3.h)) #Page 551 of Text
print('8s3 (Wdot_c/mdot)   = {}'.format(Wd_c_md2))
print('           ')
Wd_c_md3 = (st_1.cp*((T_1-T_2)+((T_2-T_3)))) #Page 551 of Text
print('8s3s (Wdot_c/mdot)   = {}'.format(Wd_c_md3))
Qd_in_md3 = (st_1.cp*(T_3-T_2))
print('cp (Qdot_in/mdot)    = {}'.format(Qd_in_md3))
#_____________________________________________________________________ Line
#_______________________Back Work Ratio ______________________________
#_____________________________________________________________________
BWR = (Wd_c_md/Wd_t_md)
BWR2 = (Wd_c_md3/Wd_t_md2)
print('1 BWR = {}'.format(BWR))
print('cp BWR = {}'.format(BWR2))
print('           ')
#_____________________________________________________________________ Line
#_____________________η Efficiency Calculations ______________________
#_____________________________________________________________________
η_t = (st_5.h-st_6.h)/(st_5.h-st_6s.h) #Original
η_CP = ((st_1.cp*(T_1-T_2))+(st_1.cp*(T_2-T_3))+(st_1.cp*(T_5-T_6)))/(st_1.cp*(T_5-T_4))
print('1. η_t  = {}'.format(η_t))
print('CP η = {}'.format(η_CP))
print('           ')
η_c1 = ((st_2.h-st_1.h)/(st_2s.h-st_1.h)) #Inversed from original
η_c1e = (st_1.cp*(T_1-T_2))

η_c2 = ((st_3s.h-st_8s.h)/(st_3.h-st_8s.h)) 
η_c2b = ((st_3.h-st_8s.h)/(st_3s.h-st_8s.h)) #Inversed from original
η_c2c = ((st_3s.h-st_8.h)/(st_3.h-st_8.h)) 
η_c2c = ((st_3.h-st_8.h)/(st_3s.h-st_8.h)) #Inversed from original
η_c2e = (st_1.cp*(T_2-T_3))

η_c  = ((η_c1+η_c2)/2)
η_cc  = ((η_c1+η_c2c)/2)
η_cb  = ((η_c1+η_c2b)/2)
η_ce  = ((η_c1e+η_c2e)/2)

print('1. η_c1  = {}'.format(η_c1))
print('CP η_c1  = {}'.format(η_c1e))
print('           ')
print('η_c2  = {}'.format(η_c2))
print('η_c2b = {}'.format(η_c2b))
print('η_c2e = {}'.format(η_c2e))
print('           ')
print('η_c  = {}'.format(η_c))
print('η_cb = {}'.format(η_cb))
print('           ')
print('η_ce = {}'.format(η_ce))
print('η_c = {}'.format(η_c))
print('           ')

η = ((Wd_cyc)/Qd_in_md)  # prints kilojoule / joule
cp = ((Wd_cyc)/Qd_in_md3) # prints kilojoule / joule
η = ((Wd_cyc)/Qd_in_md)
ηW = (1-(st_1.T/st_2.T)) #Wiki
ηW2 = ((1-(st_1.T/st_2.T))+(1-(st_2.T/st_3.T)))/2 #Wiki
print(('η       = {}'.format(η)))
print(('η_cp    = {}'.format(cp)))
print('           ')
print(('ηW      = {}'.format(ηW)))
print(('ηW2     = {}'.format(ηW2)))
print('           ')

#Find the Thermal Efficiency 
#Find the Back Work Ratio

#Find the Isentropic Efficiency of the Turbine
#How long will these efforts be studied before used for power generation.
#____________________________________________________________ Line 

print('           ')
print('           ')
print('           ')
print('h1    = {}'.format(st_1.h))
print('h2    = {}'.format(st_2.h))
print('h2s   = {}'.format(st_2s.h))
print('h3    = {}'.format(st_3.h))
print('h3s   = {}'.format(st_3s.h))
print('h8    = {}'.format(st_8.h))
print('h8s   = {}'.format(st_8s.h))
print('           ')
print('           ')
print('h4    = {}'.format(st_4.h))
print('h4s   = {}'.format(st_4s.h))
print('h5    = {}'.format(st_5.h))
print('cp1   = {}'.format(st_1.cp))
print('cp4   = {}'.format(st_4.cp))
print('           ')
print('           ')
print('1s   = {}'.format(st_1.s))
print('2s   = {}'.format(st_2.s))
print('2Ss  = {}'.format(st_2s.s))
print('3s   = {}'.format(st_3.s))
print('3Ss  = {}'.format(st_3s.s))
print('4s   = {}'.format(st_4.s))
print('4Ss  = {}'.format(st_4s.s))
print('5s   = {}'.format(st_5.s))
print('6s   = {}'.format(st_6.s))
print('6Ss  = {}'.format(st_6s.s))
print('7s   = {}'.format(st_7.s))
print('8s   = {}'.format(st_8.s))
print('8Ss  = {}'.format(st_8s.s))
print('           ')
print('           ')
print('1T    = {}'.format(st_1.T))
print('2T    = {}'.format(st_2.T))
print('2Ts   = {}'.format(st_2s.T))
print('3T    = {}'.format(st_3.T))
print('3Ts   = {}'.format(st_3s.T))
print('4T    = {}'.format(st_4.T))
print('4Ts   = {}'.format(st_4s.T))
print('5T    = {}'.format(st_5.T))
print('6T    = {}'.format(st_6.T))
print('6Ts   = {}'.format(st_6s.T))
print('7T    = {}'.format(st_7.T))
print('8T    = {}'.format(st_8.T))
print('8Ts   = {}'.format(st_8s.T))


```
